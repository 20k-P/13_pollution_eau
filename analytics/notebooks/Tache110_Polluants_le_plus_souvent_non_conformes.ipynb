###  I. Etudier le nombre de prélèvements en non conformité 

**Quels sont les prélévelment qui dépassent la limite de qualité ?**

Dans le document PDF "eau-distrib-documentation"
 
#### 1.2.14 CONFORMITE DU PRELEVEMENT
La conformité du prélèvement est définie par rapport à deux types d’exigences de qualité, les limites de qualité et les références de qualité (cf § « les exigences de qualité » dans la partie « Contexte et description des données »), en distinguant pour chaque type d’exigence les paramètres microbiologiques des substances chimiques.

##### 1.2.14.1 plvconformitebacterio
Indicateur de la conformité des paramètres microbiologiques aux limites de qualité en vigueur au moment du prélèvement pour le type d’eau considéré.
Valeurs possibles : "blanc", "C=conforme", "N=non conforme", "S (sans objet lorsqu'aucun paramètre microbio n'a été mesuré)".

##### 1.2.14.2 plvconformitechimique
Indicateur de la conformité des paramètres chimiques aux limites de qualité en vigueur au moment du prélèvement pour le type d’eau considéré (et en prenant en compte les dérogations éventuelles en cours pour l'installation concernée).
Valeurs possibles : "blanc", "C=conforme", "N=non conforme", "D=conforme dans le cadre d’une dérogation","S (sans objet lorsqu'aucun paramètre chimique n'a été mesuré)".

##### 1.2.14.2.1 plvconformitereferencebact
Indicateur de la conformité des paramètres microbiologiques aux références de qualité en vigueur au moment du prélèvement pour le type d’eau considéré.
Valeurs possibles : "blanc", "C=conforme", "N=non conforme", "S (sans objet lorsqu'aucun paramètre microbio n'a été mesuré)".

##### 1.2.14.3 plvconformitereferencechim
Indicateur de la conformité des paramètres chimiques aux références de qualité en vigueur au moment du prélèvement pour le type d’eau considéré.
Valeurs possibles : "blanc", "C=conforme", "N=non conforme", "S (sans objet lorsqu'aucun paramètre chimique n'a été mesuré)".


```python
# [TO RUN] Compter les prélévements non conformes en 2024
query_2024_count_conformity = f"""
SELECT
  COUNT(*) AS nb_prel_tot,
  SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0 END ) AS nb_prel_non_conformes,
  ROUND(SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0  END )/ COUNT(*) 
        * 100,
        2) AS percent_prel_non_conformes_vs_all,
FROM edc_prelevements
WHERE
  dateprel >= '{date_min}'
"""

con.sql(query_2024_count_conformity).show()
```

###  II. Etudier la répartition géographique de ces prélèvements non conformes.

#### 1. Quelles sont les communes avec le plus de dépassements? 

Dans le document PDF "eau-distrib-documentation"

**1.2.2 INSEECOMMUNEPRINC**<br>
Code INSEE de la commune où se trouve le point de surveillance sur lequel le prélèvement a eu lieu.

**1.2.3 NOMCOMMUNEPRINC**<br>
Nom de la commune où se trouve le point de surveillance sur lequel le prélèvement a eu lieu.

```python
#  [TO RUN] Répartir les prélévements non conformes en 2024 par communes

query_2024_count_conformity_by_com = f"""
SELECT
  NOMCOMMUNEPRINC, 
  INSEECOMMUNEPRINC,
  COUNT(*) AS nb_prel_tot,
  SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0 END ) AS nb_prel_non_conformes,
  ROUND(SUM(CASE 
        WHEN ("plvconformitebacterio" = 'N' OR  "plvconformitechimique" = 'N' OR  "plvconformitereferencebact" = 'N' OR  "plvconformitereferencechim" = 'N' )
        THEN 1 ELSE 0  END )/ COUNT(*) 
        * 100,
        2) AS percent_prel_non_conformes
FROM edc_prelevements
WHERE
  dateprel >= '{date_min}'
GROUP BY 
   NOMCOMMUNEPRINC, 
   INSEECOMMUNEPRINC
ORDER BY 
   percent_prel_non_conformes DESC,
   nb_prel_non_conformes  DESC
"""

con.sql(query_2024_count_conformity_by_com).show()
```

```python
#  [TO RUN] Répartir les prélévements non conformes en 2024 par commune

query_2024_conformity_com_repartition = f"""
WITH COMMUNES AS (
    {query_2024_count_conformity_by_com}
    ),
    
    TOTAL AS (
    SELECT
        COUNT(INSEECOMMUNEPRINC) AS tot_commune
    FROM
        COMMUNES
    )
    
SELECT
    (CASE WHEN percent_prel_non_conformes = 100 THEN '100%'
         WHEN percent_prel_non_conformes <10 THEN '0-10%'
         WHEN percent_prel_non_conformes <20 THEN '11-20%'
         WHEN percent_prel_non_conformes <30 THEN '21-30%'
         WHEN percent_prel_non_conformes <40 THEN '31-40%'
         WHEN percent_prel_non_conformes <50 THEN '41-50%'
         WHEN percent_prel_non_conformes <60 THEN '51-60%'
         WHEN percent_prel_non_conformes <70 THEN '61-70%'
         WHEN percent_prel_non_conformes <80 THEN '71-80%'
         WHEN percent_prel_non_conformes <90 THEN '81-90%'
         WHEN percent_prel_non_conformes <100 THEN '91-99%'
         END) AS percent_prel_non_conformes_cat, 
    COUNT(INSEECOMMUNEPRINC) AS nb_communes,
    MAX(tot_commune) AS tot_commune,
    ROUND(COUNT(INSEECOMMUNEPRINC)/MAX(tot_commune) *100,2) AS percent_communes
FROM
    COMMUNES
CROSS JOIN 
    TOTAL
GROUP BY
    percent_prel_non_conformes_cat
ORDER BY 
    percent_prel_non_conformes_cat
"""

con.sql(query_2024_conformity_com_repartition).show()
```

```python
# [TO RUN]

query_polluant = """ 
SELECT
  referenceprel,
  libmajparametre,
  (CASE
    WHEN
        (LOWER(libmajparametre) LIKE '%azine%' OR
        LOWER(libmajparametre) LIKE '%phos%' OR
        LOWER(libmajparametre) LIKE '%triazine%' OR
        LOWER(libmajparametre) LIKE '%glyphosate%' OR
        LOWER(libmajparametre) LIKE '%atrazine%' OR
        LOWER(libmajparametre) LIKE '%metolachlor%') THEN 'pesticides'
    WHEN
        (LOWER(libmajparametre) LIKE '%nitrate%' OR
        LOWER(libmajparametre) LIKE '%nitrite%') THEN 'nitrates et nitrites'
    WHEN
        (LOWER(libmajparametre) LIKE '%fluoro%' OR
        LOWER(libmajparametre) LIKE '%perfluoro%' OR
        LOWER(libmajparametre) LIKE '%pfoa%' OR
        LOWER(libmajparametre) LIKE '%pfos%') THEN 'PFAS'
    WHEN
        (LOWER(libmajparametre) LIKE '%phtalate%' OR
        LOWER(libmajparametre) LIKE '%dehp%' OR
        LOWER(libmajparametre) LIKE '%bbp%' OR
        LOWER(libmajparametre) LIKE '%dbp%') THEN 'phtalates'
    WHEN
        libmajparametre = 'chlorure de vinyle' THEN 'chlorure de vinyle'
    WHEN
        (LOWER(libmajparametre) LIKE '%benzène%' OR
        LOWER(libmajparametre) LIKE '%toluène%' OR
        LOWER(libmajparametre) LIKE '%naphtalène%' OR
        LOWER(libmajparametre) LIKE '%xylène%') THEN 'hydrocarbures'
    WHEN
        (LOWER(libmajparametre) LIKE '%mercure%' OR
        LOWER(libmajparametre) LIKE '%plomb%' OR
        LOWER(libmajparametre) LIKE '%arsenic%' OR
        LOWER(libmajparametre) LIKE '%cadmium%' OR
        LOWER(libmajparametre) LIKE '%nickel%' OR
        LOWER(libmajparametre) LIKE '%chrome%') THEN 'métaux lourds'
    ELSE 'other'
  END ) AS cat_polluant
FROM  
    edc_resultats 
"""

con.sql(query_polluant).show()
```

```python
# [TO RUN] Compter les prélévements non conformes en 2024 par catégorire de polluants

query_2024_count_conformity_cat = f"""
WITH
    RESULTATS AS (
        {query_polluant}
    ),

    PRELEVEMENTS AS (
        SELECT DISTINCT
            referenceprel,
            (CASE WHEN (edc_prelevements.plvconformitebacterio = 'N' OR  edc_prelevements.plvconformitechimique = 'N' 
                        OR  edc_prelevements.plvconformitereferencebact = 'N' OR  edc_prelevements.plvconformitereferencechim = 'N' )
            THEN 1 ELSE 0 END ) AS is_non_conforme,
        FROM
            edc_prelevements
        WHERE
          dateprel >= '{date_min}' 
    ),

    JOINTURE AS (
        SELECT
          RESULTATS.libmajparametre AS substance,
          RESULTATS.cat_polluant AS cat_polluant,
          RESULTATS.referenceprel AS referenceprel,
          PRELEVEMENTS.is_non_conforme AS is_non_conforme,
        FROM  
            RESULTATS  
        LEFT JOIN 
            PRELEVEMENTS
        ON 
          RESULTATS.referenceprel = PRELEVEMENTS.referenceprel
    )
  
    SELECT
      cat_polluant ,
      COUNT(*) AS nb_prel_tot,
      SUM(CASE WHEN is_non_conforme =1 THEN 1 ELSE 0 END ) AS nb_prel_non_conformes,
      ROUND(SUM(CASE WHEN is_non_conforme =1THEN 1 ELSE 0 END  )/ COUNT(*)  * 100,2) AS percent_prel_non_conformes_vs_all,
    FROM 
        JOINTURE
    GROUP BY 
        cat_polluant 
"""

con.sql(query_2024_count_conformity_cat).show()
```
